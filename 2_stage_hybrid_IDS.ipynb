```python
# 1️⃣ Load binary classifier
import numpy as np
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import tensorflow as tf
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Load binary classifier
rf_model = joblib.load("random_forest_cicids_binary.pkl")

# 2️⃣ Load multiclass dataset and preprocess features
print("\n📥 Loading multiclass dataset...")
df_multi = pd.read_csv("CICIDS2017_2018_Merged_Fuzzy_multiclass.csv", low_memory=False)
df_multi = df_multi.dropna(subset=["Label"])

X = df_multi.drop(columns=["Label"])
y_multi = df_multi["Label"]

# Clean and scale features
X = X.apply(pd.to_numeric, errors='coerce')
X = X.replace([np.inf, -np.inf], np.nan).dropna()
y_multi = y_multi.loc[X.index]  # Align labels with cleaned features

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# 3️⃣ Apply binary model to predict malicious samples
binary_preds = rf_model.predict(X_scaled)
attack_indices = np.where(binary_preds == 1)[0]
X_attack = X_scaled[attack_indices]
y_attack = y_multi.iloc[attack_indices]

# Encode multiclass labels
y_attack = y_attack.astype(str)
le = LabelEncoder()
y_attack_encoded = le.fit_transform(y_attack)
y_attack_cat = to_categorical(y_attack_encoded)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_attack, y_attack_cat, test_size=0.2, random_state=42)

# Reshape for LSTM
X_train_lstm = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_lstm = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# 4️⃣ Train LSTM model
print("\n🧠 Training LSTM model...")
lstm_model = Sequential([
    LSTM(64, input_shape=(1, X_train.shape[1]), return_sequences=False),
    Dropout(0.3),
    Dense(y_attack_cat.shape[1], activation='softmax')
])

lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
lstm_model.fit(X_train_lstm, y_train, epochs=10, batch_size=64, validation_data=(X_test_lstm, y_test))

# Extract LSTM features
print("\n🔎 Extracting LSTM features...")
lstm_feat_extractor = tf.keras.Model(inputs=lstm_model.input, outputs=lstm_model.layers[-2].output)
X_train_feat = lstm_feat_extractor.predict(X_train_lstm)
X_test_feat = lstm_feat_extractor.predict(X_test_lstm)

# 5️⃣ Train XGBoost on extracted LSTM features
print("\n🚀 Training XGBoost on LSTM features...")
xgb_model = XGBClassifier(n_estimators=100, random_state=42, use_label_encoder=False, eval_metric='mlogloss')
xgb_model.fit(X_train_feat, np.argmax(y_train, axis=1))

# 6️⃣ Save models
print("\n💾 Saving models...")
lstm_model.save("stage2_lstm_model.h5")
joblib.dump(xgb_model, "stage2_xgboost_model.pkl")
joblib.dump(le, "stage2_label_encoder.pkl")

# 7️⃣ Evaluate the stage 2 model
y_test_labels = np.argmax(y_test, axis=1)
y_pred_labels = xgb_model.predict(X_test_feat)

accuracy = accuracy_score(y_test_labels, y_pred_labels)
precision = precision_score(y_test_labels, y_pred_labels, average='weighted')
recall = recall_score(y_test_labels, y_pred_labels, average='weighted')
f1 = f1_score(y_test_labels, y_pred_labels, average='weighted')

print(f"\n✅ Accuracy: {accuracy:.4f}")
print(f"✅ Precision: {precision:.4f}")
print(f"✅ Recall: {recall:.4f}")
print(f"✅ F1 Score: {f1:.4f}")

print("\n📋 Classification Report:")
print(classification_report(y_test_labels, y_pred_labels, target_names=le.classes_))

# Confusion Matrix
cm = confusion_matrix(y_test_labels, y_pred_labels)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Confusion Matrix - Stage 2")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Metrics bar chart
metrics = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1
}

plt.figure(figsize=(8, 5))
sns.barplot(x=list(metrics.keys()), y=list(metrics.values()), palette="viridis")
plt.ylim(0, 1)
plt.title("Stage 2 Model Evaluation Metrics")
plt.ylabel("Score")
plt.show()
```

```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import classification_report, accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from xgboost import XGBClassifier

# 1️⃣ Load Merged Multiclass Dataset
df_multi = pd.read_csv("CICIDS2017_2018_Merged_Fuzzy_multiclass.csv")
df_multi = df_multi.dropna(subset=['Label'])

# 2️⃣ Preprocess Features
X = df_multi.drop(columns=['Label'])
y_multi = df_multi['Label']

X = X.apply(pd.to_numeric, errors='coerce')
X = X.replace([np.inf, -np.inf], np.nan).dropna()
y_multi = y_multi.loc[X.index]  # Align y with cleaned X

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# 3️⃣ Load Stage 1 Binary Model
rf_model = joblib.load("random_forest_cicids_binary.pkl")

# 4️⃣ Predict Benign vs Attack
binary_preds = rf_model.predict(X_scaled)
attack_indices = np.where(binary_preds == 1)[0]

# 5️⃣ Filter Only Malicious Samples
X_attack = X_scaled[attack_indices]
y_attack = y_multi.iloc[attack_indices]

# 6️⃣ Encode Labels
le = LabelEncoder()
y_attack_encoded = le.fit_transform(y_attack)
y_attack_cat = to_categorical(y_attack_encoded)

# 7️⃣ Split and Reshape for LSTM
X_train, X_test, y_train, y_test = train_test_split(X_attack, y_attack_cat, test_size=0.2, random_state=42)

X_train_lstm = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_lstm = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# 8️⃣ Build LSTM Model
lstm_model = Sequential([
    LSTM(64, input_shape=(1, X_train.shape[1]), return_sequences=False),
    Dropout(0.3),
    Dense(y_attack_cat.shape[1], activation='softmax')
])

lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 9️⃣ Train LSTM
print("🚀 Training LSTM...")
history = lstm_model.fit(X_train_lstm, y_train, epochs=10, batch_size=64, validation_data=(X_test_lstm, y_test))

# 🔟 Extract LSTM Features
_ = lstm_model.predict(X_train_lstm[:1])  # Trigger model to build
lstm_feat_extractor = Model(inputs=lstm_model.input, outputs=lstm_model.layers[-2].output)

X_train_feat = lstm_feat_extractor.predict(X_train_lstm)
X_test_feat = lstm_feat_extractor.predict(X_test_lstm)

# 1️⃣1️⃣ Train XGBoost on LSTM Features
xgb_model = XGBClassifier(n_estimators=100, random_state=42)
xgb_model.fit(X_train_feat, np.argmax(y_train, axis=1))

# 1️⃣2️⃣ Evaluate XGBoost
y_pred = xgb_model.predict(X_test_feat)
y_true = np.argmax(y_test, axis=1)

print("\n📊 Classification Report:\n", classification_report(y_true, y_pred, target_names=le.classes_))
print("✅ Accuracy:", accuracy_score(y_true, y_pred))
print("✅ F1 Score:", f1_score(y_true, y_pred, average='weighted'))
print("✅ Precision:", precision_score(y_true, y_pred, average='weighted'))
print("✅ Recall:", recall_score(y_true, y_pred, average='weighted'))

# 1️⃣3️⃣ Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()

# 1️⃣4️⃣ Save Models
lstm_model.save("stage2_lstm_model.h5")
joblib.dump(xgb_model, "stage2_xgboost_model.pkl")
joblib.dump(le, "stage2_label_encoder.pkl")
print("✅ Stage 2 models saved.")
```
